## **3. Chargement des données**
Nous allons maintenant charger notre dataset .

```python
# chargement du dataset
data = pd.read_csv('../data/data.csv')

# affichage d'un extrait
data.head()
```

## **4. Structure et typologie des données**
Nous allons analyser la structure de notre jeu de données et le types des données et changer les types de données si nécessaires.

```python
# dimension du dataset
data.shape
```

```python
# affichage des infos sur les variables et leurs types
data.info()
```

Dans l'affichage de l'extrait du dataframe dans `1.3` , nous avons remarquer que les valeurs manquantes étaient remplacées par `?`. Donc nous allons remplacer `?` par `np.nan`

```python
# remplacement de ? par np.nan
data.replace('?',np.nan,inplace=True)
```

Nous remarquons que tous les types devraient être des types numeriques donc nous allons les convertir.

```python
# conversion des types object ou str en float ou int
data = data.apply(pd.to_numeric, errors="coerce")
data.dtypes
```

## **5. Préparation et nettoyage préliminaire**

### **5.1. Gestion des valeurs manquantes et des valeurs nulles**

```python
# verification des valeurs nulles
data.isna().sum()
```

```python
# les variables slope , ca et that ont trop de valeurs maquantes , nous allons créer un nouveau dataset sans elles.
df = data.copy().drop(columns=['slope','ca','thal'])
df.head()
```

```python
# Nous allons attribuer aux valeurs manquantes de chol qui representent 7% du dataset la médiane , plus robuste que la moyenne
mediane =df['chol'].median()
df.fillna({'chol':mediane},inplace=True)
df['chol'].isna().sum()
```

```python
# on peut supprimer les lignes avec des valeurs manquantes qui representent très peu dans le dataset
df=df.dropna()
```

```python
# verification des valeurs manquantes
df.isna().sum()
```

```python
# verification des valeurs nulles
df.isnull().sum()
```

### **5.2. Gestion des doublons**

```python
# verification des valeurs dupliquées
df.duplicated().sum()
```

```python
# affichage des valeurs dupliquées
df[df.duplicated(keep=False)]
```

Dans notre contexte nous allons garder ces valeurs dupliquées car on a pas de colonne comme identifiant unique donc ces lignes pourraient concerner des patients qui sont distincts mais avec des valeurs identiques.

### **5.3. Detection et gestion des outliers**

```python
# affichage des statistiques descriptives
df[['age','trestbps','chol','thalach','oldpeak']].describe()
```

Nous remarquons que le maximum au niveau de `chol` est assez élevé donc nous allons investiguer en utilisant la méthode de l'IQR

```python
# calcul de l'IQR
q1,q3 = np.quantile(df['chol'],q=[0.25,0.75])

# limite basse
lb = q1 - 1.5 * (q3-q1)

# limite haute
lh = q3 + 1.5 * (q3-q1)

print(f"{lb,lh=}")
```

```python
df[df.chol<lb]
```

```python
df[df.chol>lh]
```

Bien que ces valeurs soient extrêmes dans notre dataset, ils sont biologiquement possible . Ces lignes sont aussi très prédictifs donc nous allons les garder.
